```python
def move_right(name, step, value_back):
    df = pd.read_csv(name)
    interval = df.loc[1, 'time_stamp'] - df.loc[0, 'time_stamp']
    
#     if (step % interval != 0):
#         step = step // interval * interval
#     if (value_back % interval != 0):
#         value_back = value_back // interval * interval
    
    rows = df.shape[0] - 1
    max_time = 0
    max_data = 0
    min_data = 100
    min_row = 0
    
    for x in range(0, rows + 1):           
        if df.loc[x, 'data'] > max_data:
            max_time = df.loc[x, 'time_stamp']
            max_data = df.loc[x, 'data']
    
    for x in range(0, rows + 1):    
        if (df.loc[x, 'data'] < 0.6) & (df.loc[x, 'time_stamp'] < max_time):
            min_row = x
            
    min_row = min_row - 30
    for x in range(min_row, rows + 1):
        df.loc[x, 'data1'] = df.loc[x, 'data1'] + step
#         df.loc[x, 'time_stamp'] = df.loc[x, 'time_stamp'] + step / interval
        
    count = 0
    for x in range(min_row - value_back, min_row):
        df.loc[x, 'data1'] = df.loc[x, 'data1'] + (step // value_back) * count
#         df.loc[x, 'time_stamp'] = df.loc[x, 'time_stamp'] + count * interval
        count = count + 1
#     df = df.sort_values(by='time_stamp',axis = 0,ascending = True)
        
    df2 = pd.DataFrame(columns=['time_stamp','data','data1'])
    i = 0
    lst = [0, 0, 0]
#     for x in range(0, rows - 2):
#         if df.loc[x + 1, 'time_stamp'] - df.loc[x, 'time_stamp'] > interval:
#             to_add = int((df.loc[x + 1, 'time_stamp'] - df.loc[x, 'time_stamp']) // interval)
#             total_data = df.loc[x + 1, 'data'] - df.loc[x, 'data']
#             total_data1 = df.loc[x + 1, 'data1'] - df.loc[x, 'data1']
#             ave_data = total_data / to_add
#             ave_data1 = total_data1 / to_add 
#             for y in range(0, to_add - 1):
#                 df2.loc[i] = lst
#                 df2.loc[i, 'time_stamp'] = df.loc[x, 'time_stamp'] + (y + 1) * interval
#                 df2.loc[i, 'data'] = df.loc[x, 'data'] + (y + 1) * ave_data
#                 df2.loc[i, 'data1'] = df.loc[x, 'data1'] + (y + 1) * ave_data1
#                 i = i + 1
                
#     df = pd.concat([df, df2])
    df = df.drop_duplicates(subset='time_stamp') 
    df = df.sort_values(by='time_stamp',axis = 0,ascending = True)
    
    return df
```
